```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# READ IMAGES

lena = cv2.imread('lena.png', 0)
boxes = cv2.imread('boxes.png', 0) 
```

```python
# BLUR IMAGES

lena_blur = cv2.GaussianBlur(lena, (9,9), 10, 10, cv2.BORDER_DEFAULT)
boxes_blur = cv2.GaussianBlur(boxes, (9,9), 0)
```

```python
# APPLY SOBEL OPERATOR

lena_sobel = cv2.Sobel(lena_blur, cv2.CV_64F, 1, 1, ksize=11)
boxes_sobel = cv2.Sobel(boxes_blur, cv2.CV_64F, 1, 1, ksize=11)
```

```python
# APPLY CANNY OPERATOR

lena_canny = cv2.Canny(lena_blur, 100, 200)
boxes_canny = cv2.Canny(boxes_blur, 100, 200)
```

```python
# PLOTTING IMAGES ----> ORIGINAL // SOBEL // CANNY

fig = plt.figure(figsize=(20,20))
plt.subplot(141), plt.imshow(lena, cmap='gray'), plt.axis('off')
plt.subplot(142), plt.imshow(lena_blur, cmap='gray'), plt.axis('off')
plt.subplot(143), plt.imshow(lena_sobel, cmap='gray'), plt.axis('off')
plt.subplot(144), plt.imshow(lena_canny, cmap='gray'), plt.axis('off')
plt.show()
```

```python
# PLOTTING IMAGES ----> ORIGINAL // SOBEL // CANNY

fig = plt.figure(figsize=(20,20))
plt.subplot(141), plt.imshow(boxes, cmap='gray'), plt.axis('off')
plt.subplot(142), plt.imshow(boxes_blur, cmap='gray'), plt.axis('off')
plt.subplot(143), plt.imshow(boxes_sobel, cmap='gray'), plt.axis('off')
plt.subplot(144), plt.imshow(boxes_canny, cmap='gray'), plt.axis('off')
plt.show()
```

----

### Now changing the sigma on the gaussian filter to see the effect on the canny operator

Notice that we change the sigma parameters and not the kernel size !!

**Decreasing Sigma**

```python
# BLUR IMAGES

lena_blur_dec = cv2.GaussianBlur(lena, (9,9), 1, 1, 0)
boxes_blur_dec = cv2.GaussianBlur(boxes, (9,9), 1, 1, 0)
```

```python
# APPLY SOBEL OPERATOR

lena_sobel = cv2.Sobel(lena_blur_dec, cv2.CV_64F, 1, 1, ksize=11)
boxes_sobel = cv2.Sobel(boxes_blur_dec, cv2.CV_64F, 1, 1, ksize=11)
```

```python
# APPLY CANNY OPERATOR

lena_canny = cv2.Canny(lena_blur_dec, 100, 200)
boxes_canny = cv2.Canny(boxes_blur_dec, 100, 200)
```

```python
# PLOTTING IMAGES ----> ORIGINAL // SOBEL // CANNY

fig = plt.figure(figsize=(20,20))
plt.subplot(141), plt.imshow(lena, cmap='gray'), plt.axis('off')
plt.subplot(142), plt.imshow(lena_blur_dec, cmap='gray'), plt.axis('off')
plt.subplot(143), plt.imshow(lena_sobel, cmap='gray'), plt.axis('off')
plt.subplot(144), plt.imshow(lena_canny, cmap='gray'), plt.axis('off')
plt.show()
```

```python
# PLOTTING IMAGES ----> ORIGINAL // SOBEL // CANNY

fig = plt.figure(figsize=(20,20))
plt.subplot(141), plt.imshow(boxes, cmap='gray'), plt.axis('off')
plt.subplot(142), plt.imshow(boxes_blur_dec, cmap='gray'), plt.axis('off')
plt.subplot(143), plt.imshow(boxes_sobel, cmap='gray'), plt.axis('off')
plt.subplot(144), plt.imshow(boxes_canny, cmap='gray'), plt.axis('off')
plt.show()
```

----

**Increasing Sigma**

```python
# BLUR IMAGES

lena_blur_inc = cv2.GaussianBlur(lena, (9,9), 20, 20, 0)
boxes_blur_inc = cv2.GaussianBlur(boxes, (9,9), 20, 20, 0)
```

```python
# APPLY SOBEL OPERATOR

lena_sobel = cv2.Sobel(lena_blur_inc, cv2.CV_64F, 1, 1, ksize=11)
boxes_sobel = cv2.Sobel(boxes_blur_inc, cv2.CV_64F, 1, 1, ksize=11)
```

```python
# APPLY CANNY OPERATOR

lena_canny = cv2.Canny(lena_blur_inc, 100, 200)
boxes_canny = cv2.Canny(boxes_blur_inc, 100, 200)
```

```python
# PLOTTING IMAGES ----> ORIGINAL // SOBEL // CANNY

fig = plt.figure(figsize=(20,20))
plt.subplot(141), plt.imshow(lena, cmap='gray'), plt.axis('off')
plt.subplot(142), plt.imshow(lena_blur_inc, cmap='gray'), plt.axis('off')
plt.subplot(143), plt.imshow(lena_sobel, cmap='gray'), plt.axis('off')
plt.subplot(144), plt.imshow(lena_canny, cmap='gray'), plt.axis('off')
plt.show()
```

```python
# PLOTTING IMAGES ----> ORIGINAL // SOBEL // CANNY

fig = plt.figure(figsize=(20,20))
plt.subplot(141), plt.imshow(boxes, cmap='gray'), plt.axis('off')
plt.subplot(142), plt.imshow(boxes_blur_inc, cmap='gray'), plt.axis('off')
plt.subplot(143), plt.imshow(boxes_sobel, cmap='gray'), plt.axis('off')
plt.subplot(144), plt.imshow(boxes_canny, cmap='gray'), plt.axis('off')
plt.show()
```

----

#### Conclusion  
as we decrease the sigma parameter in the gaussian filter, the edges with the sobel and canny operators are more clear, while when we increase said parameter, the edges tend to "disappear" and not be quite as visible.
